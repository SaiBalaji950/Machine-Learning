```python
FP Growth

FP Growth (Frequently Pattern Growth) is an efficient algorithm for mining frequent itemsets 
without candidate generations. It works by constructing compact data structure is called 
FP Growth which is generally faster than the apriori algorithm, especially for larger datasets.
```

```python
Steps of the FP-Growth Algorithm
Construct the FP-tree:

Scan the Database: Count the frequency of each item in the dataset. Filter out infrequent items based on a minimum support threshold.
Sort Items: For each transaction, sort the items in decreasing order of their frequency.

    
Build the FP-tree:


Start with a root node (null).
For each transaction, add the sorted items to the tree:
If the path exists, increment the count of the nodes in the path.
If the path doesn't exist, create new nodes along the path.


Mining the FP-tree:

Extract Frequent Patterns:
For each item in the FP-tree (starting from the least frequent), construct a conditional pattern base, 
which is a sub-database consisting of the prefixes of paths ending with that item.

From the conditional pattern base, construct the conditional FP-tree.
Recursively mine the conditional FP-tree for frequent itemsets.

Combine Results: Combine all the frequent itemsets found from the FP-trees to form the final set of frequent itemsets.
```

```python
import pandas as pd
a=pd.read_csv('diabetes.csv')
display(a)
```

```python
one_hot_data=pd.get_dummies(a)
one_hot_data = one_hot_data.astype(bool)
```

```python
from mlxtend.frequent_patterns import fpgrowth,association_rules              #pip install mlxtend
```

```python
c=fpgrowth(one_hot_data,min_support=0.003,use_colnames=True)
d=association_rules(c,metric='confidence',min_threshold=0.2)
print("FP Growth:\n",c)
print("\nAssociation Rules:\n",d)
```
