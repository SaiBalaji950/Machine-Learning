```python
Apriori Algorithm

Apriori Algorithm uses frequent itemset to generate the association rule, and it is designed to work
on the databases that contains Transactions. with the help of these association rule, that determines 
how strongly or how weakly two objects are connected. this algorithm uses the Breadth-first-search and
hash tree to calculate the itemset associations efficiently. It is an iterative process for finding the
frequent itemsets from the large datasets.
```

```python
Support

Support is the proportion of the transactions in a dataset that contain a specific itemset.
it indicates how frequently an itemset appears in the dataset.

Support Formula:

Support(X)=(Number of transactions containing X)/(Total No.of Transactions)
```

```python
Confidence

Confidence is a measure of how often items in a rule appear together. It indicates 
the likelihood of finding the consequent(B) in transactions that contains antecedent(A).

Confidence(A‚ÜíB)=Support(A‚à™B)/ Support(A)
```

```python
Lift is a measure used in association rule mining to evaluate the strength of a rule 
ùê¥
‚Üí
ùêµ
A‚ÜíB. It compares the observed support of the rule to the expected support if 
ùê¥
A and 
ùêµ
B were independent. A lift value greater than 1 indicates a positive association between 
ùê¥
A and 
ùêµ
B, while a lift less than 1 indicates a negative association.

Definition
Lift is calculated as follows:

Lift(ùê¥‚Üíùêµ)=Support(ùê¥‚à™ùêµ)/(Support(ùê¥)√óSupport(ùêµ))
```

```python
What is Frequent Itemset

Frequent Itemset are those items whose support greater than threshold value or user specified 
minimum value. it means A & B are frequent itemset each other, then individually A & B both are  
also should be freququent itemset.

Supoose there are two frequent itemset,
A={1,2,3,4,5}
B={2,3,7}
From above two transactions, 2,3 are frequent itemset.
```

```python
Steps for Apriori Algorithm

1) Determine the support of itemsets in the transactional databases and then select 
    the minimum support and confidence.
2) Select all the support in the transaction with higher support value than the 
    minimum or selected support value.
3) Find all the rules of these subsets that have higher confidence value then threshold
    or minimum confidence.
4) Sort the rules as the decreasing order of lift.
```

```python
import pandas as pd
a=pd.read_csv('diabetes.csv')
display(a)
```

```python
print(a.shape)
```

```python
transactions=[]
for i in range(0,768):
    transactions.append([str(a.values[i,j]) for j in range(0,9)])
```

```python
from apyori import apriori
```

```python
rules=apriori(transactions=transactions,min_support=0.003,min_confidence=0.2,min_left=3,min_length=2,max_length=2)
```

```python
results=list(rules)
print(results)
```
