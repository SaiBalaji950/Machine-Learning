```python
Polynomial Regression

The Regression Algorithm that models the relationship between the independent variable(x) and dependent variable(y) as nth degree polynomial is 
called Polynomial Regression.

Equation for Polynomial Regression is y=b0+b1x+b2x^2+b3x^3+......+bnx^n
```

```python
1) It is also called the special case of Multiple Linear Regression. Because we add multiple Polynomial terms to
    Multiple Linear Regression equation to convert into Polynomial Regression.
2) It is a linear model with some modification in order to increase the accuracy.
3) The Dataset used in Polynomial Regression for training is of non-linear nature.
4) it makes use of a linear regression model to fit the complicated and non-linear functions and datasets.
5) Hence in Polynomial Regression, the original features are converted into Polynomial features of required degree(2,3,..n) and then modeled using in a
    linear model.
```

```python
Need for Polynomial Regression:

1) When we apply a linear model on a linear dataset,it will provides good result as we have seen in the simple Linear Regression.But when we apply 
   a linear model on a non-linear datset,it will provides the drastic result. Due to which loss function and error rate increases and accuracy rate 
   will decreases.
2) So for such cases, where datapoints are arranged in a non-linear function,we use Polynomial Regression model.
```

```python
Non-Linear Regression

Non-Linear Regression is a model used to the relationship between independent variable and dependent variable when the data 
cannot be accurately represented by a linear model. unlike linear regression,which assumes a straight-line relationship,non-linear 
regression models accomodate curves and more complex patterns.

Types of Non-Linear Regression Models:

1) Polynomial Regression
2) Exponential Regression
3) Power Law Regression
4) Logistic Regression
5) Logarithmic Regression
```

```python
we can't import Non-Linear Regression directly. we want to implement the polynomial Regression.
Polynomail Regression,which can model the non-linear relationships by extending the linear model to include polynomial terms.

And also we used pipeline for multiple preprocessing and model training in a single workflow.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
a=pd.read_csv('output.csv')
b=LabelEncoder()
a['names']=b.fit_transform(a['names'])
x=a[['names','id']]
y=a['marks']
c=ColumnTransformer([('encoder',OneHotEncoder(),[0])],remainder='passthrough')
c.fit_transform(a)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
d=LinearRegression()
d.fit(x_train,y_train)
y_pred=d.predict(x_test)
x_pred=d.predict(x_train)
print("Linear Regression score is")
print(d.score(x_train,y_train))
print(d.score(x_test,y_test))
poly=PolynomialFeatures(degree=3)
x_poly=poly.fit_transform(x)
d=LinearRegression()
d.fit(x_poly,y)
pipeline=Pipeline([('PreProcessor',c),('Polynomial',poly),('Linear_Regression',d)])
pipeline.fit(x_train,y_train)
y_pred=pipeline.predict(x_test)
x_pred=pipeline.predict(x_train)
print("Pipeline score is")
print(pipeline.score(x_train,y_train))
print(pipeline.score(x_test,y_test))
plt.plot(x,y,'*:r')
plt.title('My creation')
plt.xlabel('names & id')
plt.ylabel('marks')
plt.legend(x)
plt.show()
```

```python
Bayesian Linear Regression

Bayesian Linear Regression is an approach for the linear regression that incorporates the bayesian principle to estimate the 
parameters of the model.Unlike Traditional Linear Regression,which provides a single estimates of the model parameters,
Bayesian Linear Regression provides a distribution over the possible parameter values, reflecting the uncertainity in these estimates.
```

```python
Bayesian Framework

Prior Distribution: Represents the initial belief about the parameters before observing any data.
                    For Linear Regression,this is usually a gussian distribution centered about zero.
Likelihood: Represents how likely the observed data is given the parameters.
            For Linear Regression,this is typically assumed to the gussian noise around the linear prediction.
Posterior Distribution: Updated beliefs about the parameters after observing the data. it combines the prior distribution
                        and likelihood using Bayes theorem.
```

```python
Bayes theorem 

P(θ∣X,y)= P(y∣X)/P(y∣X,θ)⋅P(θ)
where,P(θ∣X,y) is the posterior distribution of the parameters 𝜃 given data 𝑋 and targets y.
    P(y∣X,θ) is the likelihood of the data given the parameters.
    P(θ) is the prior distribution of the parameters.
    P(y∣X) is the marginal likelihood or evidence.
```

```python
import numpy as np
from sklearn.linear_model import BayesianRidge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
x=np.random.rand(100,2)
y=3+2*x[:,0]+4*x[:,1]+np.random.randn(100)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
a=BayesianRidge()
a.fit(x_train,y_train)
y_pred=a.predict(x_test)
print(a.coef_)
print(a.intercept_)
print(mean_squared_error(y_pred,y_test))
```

```python
Regression Trees 

Regression Trees are also known as Decision Tree Regressor as a type of tree based model used for predicting numerical values.
```

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
x=np.random.rand(100,2)
y=3+2*x[:,0]+4*x[:,1]+np.random.randn(100)
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
a=DecisionTreeRegressor()
a.fit(x_train,y_train)
y_pred=a.predict(x_test)
print(mean_squared_error(y_pred,y_test))
```
