```python
SVM

SVM stands for Support Vector Machine.
It is one of the most popular supervised Learning Technique and used for to solve Classification and Regression problems. 
But Primarily it is used for the classification problems.

The Goal of the SVM is used to create the best line or decision boundary that can segregate(which seperates or divides something into parts or groups)
n-dimensional space into classes so that we can easily put the new data into new category in the future. Hence the best line or decision boundary is 
called Hyperplane.

SVM chooses the extreme points/vectors that help in creating the hyperplane. These Exteme cases are called as Support Vectors and algorithm 
can be defined as Support Vector Machines.
```

```python
Types of SVM

Linear SVM:
Linear SVM is used for linearly separable data, which means if dataset is classified into two classes by using a straight line , then such data 
is termed as linear separable data and classifier is used called as linear SVM.

Non-Linear SVM:
Non-Linear SVM is used for non-linear separated data, which means if dataset is not classified by using a straight line, then such datais termed 
as non-linear data and classifier is used called as non-linear SVM.
```

```python
Types of SVM

Linear SVM:
Linear SVM is used for linearly separable data, which means if dataset is classified into two classes by using a straight line , then such data 
is termed as linear separable data and classifier is used called as linear SVM.

Non-Linear SVM:
Non-Linear SVM is used for non-linear separated data, which means if dataset is not classified by using a straight line, then such datais termed 
as non-linear data and classifier is used called as non-linear SVM.
```

```python
Hyperplane 

There can be multiple best lines/decision boundary which can segregate n-dimensional space into classes but we need to find out the
best decision boundary that helps to classify the new data points. This Best Boundary is known as the Hyperplane.

Support Vectors
The data points are or closest to the hyperplane and which may affect the position of the hyperplane are termed as support vectors.
since these vectors support the hyperplane,hence called a Support vectors.
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=pd.read_csv('diabetes.csv')
x=a.iloc[:,[2,3]].values
y=a.iloc[:,4].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
b=StandardScaler()
x_train=b.fit_transform(x_train)
x_test=b.fit_transform(x_test)
c=SVC(kernel='linear',random_state=42)
c.fit(x_train,y_train)
y_pred=c.predict(x_test)
x_pred=c.predict(x_train)
d=accuracy_score(y_test,y_pred)
e=confusion_matrix(y_test,y_pred)
f=classification_report(y_test,y_pred,zero_division=0)
print("Accuracy:",d)
print("Confusion_matrix:\n",e)
print("classification_report:\n",f)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
a=pd.read_csv('diabetes.csv')
x=a[['Pregnancies']]
y=a['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
pipeline=Pipeline([('Scaler',StandardScaler()),('SVM',SVC(kernel='linear',probability=True,random_state=42))])
pipeline=pipeline.fit(x_train,y_train)
b=np.linspace(x.iloc[:,0].min() -1,x.iloc[:,0].max() +1,500)
b_grid=pd.DataFrame(b,columns=['Pregnancies'])
c=pipeline.predict_proba(b_grid)[:,1]
plt.scatter(b,c,color='r',label='Predicted Probability')
plt.scatter(x.iloc[:,0],y,c=y,cmap='winter_r',marker='o',edgecolor='b',label='Data Points')
plt.colorbar(label='Probability')
plt.title("Support Vector Machine")
plt.xlabel('Pregnancies')
plt.ylabel('Outcome')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=pd.read_csv('diabetes.csv')
x=a[['Pregnancies']]
y=a['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
b=StandardScaler()
x_train=b.fit_transform(x_train)
x_test=b.fit_transform(x_test)
c=SVC(kernel='linear',random_state=42)
c.fit(x_train,y_train)
y_pred=c.predict(x_test)
x_pred=c.predict(x_train)
d=accuracy_score(y_test,y_pred)
e=confusion_matrix(y_test,y_pred)
f=classification_report(y_test,y_pred,zero_division=0)
print("Accuracy:\n",d)
print("Confusion Matrix:\n",e)
print("Classification Report:\n",f)
```
