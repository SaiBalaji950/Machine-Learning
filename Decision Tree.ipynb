```python
Decision Trees

Decision Trees is a Supervised Learning Technique which is used for both classification and Regression algorithm.
But Majorly used for the classification Problems.

It is a tree-structured classifier where internal nodes represent the features of the dataset and branches represent the decision rules
and each node represents the outcome.

In Decision Trees there are two nodes, Decision nodes and leaf nodes. decision nodes are used to make any decisions and have multiple branches.
but leaf nodes are output of those decisions and cannot have any further branches.

The decisions or test are performed on the basis of features of the given dataset.
It is a graphical representation for getting all possible solutions to a problem/solutions based on the given conditions.
It is called Decision Tree Because it is similar to a tree it starts with a root node and expands into branches and constructs tree-like structure.

In order to build a tree we use CART Algorithm - Classification And Regression Algorithm.
Decision Tree is simply asks a question based on the answer Yes or No it further split the tree into subtrees.
```

```python
Decision Tree Terminologies

Root Node: Root node is from where the decision tree starts. It Represents the entire datasets, which further gets divided into two 
or more homogeneous sets.

Leaf Node: Leaf Node are the final output node and the tree cannot be segregated further after getting the leaf node.

Splitting: Splitting is the process of dividing decision nodes/root nodes into sub nodes according to the given conditions.

Branch/Sub Trees: A Tree formed by the splitting Tree.

Pruning: Pruning is the process of remove the unwanted branches from the tree.

Parent/Child Node: The Root node of the tree is called Parent Node and other nodes are called child nodes.
```

```python
Working of the Decision Tree:

1) Begin the tree with the root node, says S, which contains the complete dataset.
2) Find the Best Attribute in the dataset by using Attribute Selection Measure(ASM).
3) Divides the S into subsets that contains possible values for the best attributes.
4) Generate the Decision Tree node, until we got the best attribute.
5) Recursively make new decision trees using the subsets of the dataset created in step-3. continue the process until the stage is reacher where
    we can't classified the nodes are called the final node as a leaf node.
```

```python
Attribute Selection Measure

In the Decision Tree While we selecting the best attribute sometimes there are some problems will be raised.
to solve the problems we used Attribute Selection Measure(ASM). By this Measurement we can easily solve the problems and we can find Best attribute.

There are two techniques in ASM- Information Gain and Gini Index.
```

```python
Information Gain

Before Going to Information Gain We Want to know about the Entopy.

Entropy is a metric to measure the impurities in the given dataset. it specifies the Randomness. Entropy can be calculated as:

Entropy (s)= -P(Yes) log2 P(Yes) - P(No) log2 P(No)

where s= Total Number of Samples
     P(Yes)=Probability of Yes
     P(No)=Probability of No

Now Information Gain is 

Information Gain is the measurement of changes in the entropy after the segmentation of the dataset based on the attribute.
It Calculates how much information a feature provides us about a class.
According to the value of information gain, we split the node and build the decision tree.
A Decision Tree Algorithm always tries to maximize the value of information gain and add a node having the highest information gain is split first.

It can be calculated as 
Information Gain= Entropy -[Weighted Average *Entropy(All Features)]
```

```python
Gini Index

Gini index is a measure of impurity and purity of used while creating a decison tree in the CART Algorithm.
An Attribute with low gini index is always should be preferred with high gini index.
It uses only Binary split, and the CART Algorithm uses Gini index to create binary splits.
It can be calculated as 
Gini Index= 1- ∑jPj2
```

```python
Pruning: Getting an Optimal Decision tree
Pruning is a process of deleting the unnecessary nodes from a tree in order to get the optimal decision tree.

A too-large tree increases the risk of overfitting, and a small tree may not capture all the important features of the dataset. 
Therefore, a technique that decreases the size of the learning tree without reducing accuracy is known as Pruning. 

There are mainly two types of tree pruning technology used:

Cost Complexity Pruning
Reduced Error Pruning.
```

```python
Advantages of the Decision Tree

It is simple to understand as it follows the same process which a human follow while making any decision in real-life.
It can be very useful for solving decision-related problems.
It helps to think about all the possible outcomes for a problem.
There is less requirement of data cleaning compared to other algorithms.
```

```python
Disadvantages of the Decision Tree


The decision tree contains lots of layers, which makes it complex.
It may have an overfitting issue, which can be resolved using the Random Forest algorithm.
For more class labels, the computational complexity of the decision tree may increase.
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=pd.read_csv('diabetes.csv')
x=a.iloc[:,[2,3]]
y=a.iloc[:,4]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
b=StandardScaler()
x_train=b.fit_transform(x_train)
x_test=b.fit_transform(x_test)
c=DecisionTreeClassifier(criterion='entropy',random_state=0)
c.fit(x_train,y_train)
y_pred=c.predict(x_test)
x_pred=c.predict(x_train)
d=accuracy_score(y_test,y_pred)
e=confusion_matrix(y_test,y_pred)
f=classification_report(y_test,y_pred,zero_division=0)
print("Accuracy:",d)
print("Confusion_matrix:\n",e)
print("classification_report:\n",f)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
a=pd.read_csv('diabetes.csv')
x=a[['Pregnancies']]
y=a['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
pipeline=Pipeline([('Scaler',StandardScaler()),('DTC',DecisionTreeClassifier(criterion='entropy',random_state=42))])
pipeline=pipeline.fit(x_train,y_train)
b=np.linspace(x.iloc[:,0].min() -1,x.iloc[:,0].max() +1,500)
b_grid=pd.DataFrame(b,columns=['Pregnancies'])
c=pipeline.predict_proba(b_grid)[:,1]
plt.scatter(b,c,color='r',label='Predicted Probability')
plt.scatter(x.iloc[:,0],y,c=y,cmap='winter_r',marker='o',edgecolor='b',label='Data Points')
plt.colorbar(label='Probability')
plt.title("Support Vector Machine")
plt.xlabel('Pregnancies')
plt.ylabel('Outcome')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
a = pd.read_csv('diabetes.csv')
x = a[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
y = a['Outcome']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and fit the pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('dtc', DecisionTreeClassifier(criterion='entropy', random_state=42))
])
pipeline.fit(x_train, y_train)

# Define a mesh grid for plotting decision boundaries
x_min, x_max = x[['Pregnancies']].min() - 1, x[['Pregnancies']].max() + 1
b = np.linspace(x_min, x_max, 500)
b_grid = pd.DataFrame(b, columns=['Pregnancies'])

# Predict probabilities for the mesh grid
c = pipeline.predict_proba(b_grid)[:, 1]

# Plotting decision boundary and data points
plt.figure(figsize=(10, 6))
plt.plot(b, c, color='r', label='Predicted Probability')
plt.scatter(x['Pregnancies'], y, c=y, cmap='winter_r', marker='o', edgecolor='b', label='Data Points')
plt.colorbar(label='Probability')
plt.title("Decision Tree Classifier")
plt.xlabel('Pregnancies')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=pd.read_csv('diabetes.csv')
x=a[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
y=a['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
b=StandardScaler()
x_train=b.fit_transform(x_train)
x_test=b.fit_transform(x_test)
c=DecisionTreeClassifier(criterion='entropy',random_state=42)
c.fit(x_train,y_train)
y_pred=c.predict(x_test)
x_pred=c.predict(x_train)
d=accuracy_score(y_test,y_pred)
e=confusion_matrix(y_test,y_pred)
f=classification_report(y_test,y_pred,zero_division=0)
print("Accuracy:\n",d)
print("Confusion Matrix:\n",e)
print("Classification Report:\n",f)
```
