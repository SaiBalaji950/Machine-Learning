```python
Simple Linear Regression Model

If there is only one input variable then the Linear Regression is called Simple Linear Regression.

y= a0+a1x+ε 

y=Dependent variable
x=Independent variable
a0=intercept of the line(Give an additional degree of freedom)
a1=Linear Regression co-efficient(scale factor to each input value)
ε=Error term
```

```python
Assumptions

1) A Linear relationship should exist only in target and predictor variables.
2) The Regression Residuals must be normally distributed.
3) Multiple Linear Regressions assumes little or no multicollinearity in data.
```

```python
Backward Elimination

Backward Elimination is a feature selection while building a machine learning model. it is used to remove those features 
that do not have a significant effect on the dependent variable or prediction of output.

there are various methods to build a model in machine learning are:

1) All-in
2) Backward Elimination
3) Bi-directional Elimination
4) Forward Selection
5) Score Comparision

but Backward Elimination is the fastest method while comparing to four methods.
```

```python
Steps for Backward Elimination

1) Firstly we need to select a significance level to stay in the model.
2) fit the complete model with all possible predictor values.
3) choose the predictor which has the highest p-value.
4) Remove that predictor
5) Rebuild and fit the model with remaining variables.
```
