```python
Unsupervised Learning
Unsupervised Learning is a learning method in which a machine learns without any supervision.
that means Unsupervised Learning is learns the data without labels,classified or categorized and the algorithm needs to act on the data without 
any supervision.

Unsupervised Learning is also classified into two types. They are:
1) Clustering
2) Association
```

```python
Supervised Learning
In Supervised Learning, sample labelled data was provided for machine learning for training and the system then predicts the output
based on the training data.
Supervised learing is also known as labelled data.
The System uses labelled data to build a model that understands the datasets and learn about each one.
After the training and processing are done, we test the model with sample data to see it if an accurately predict the output.

Supervised Learning is also classified into two types. They are:
1) Classification
2) Regression
```

```python
How Supervised Learning Works:

Working Steps for Supervised Learning:
1) First Determine the Training Dataset.
2) Collect/Gather the labelled training data.
3) Split the dataset into training dataset, test dataset, validation test.
4) Determine the input features of the training dataset, which should have enough knowledge so that the model can accurately predict the output.
5) Determine the suitable algorithm for the model, such as decision trees, random forests, k-nearest neighbors etc.
6) Execute the algorithm on the training dataset. sometimes we need validation sets as the control parameters,which are the subset of the training 
    dataset.
7) Evaluate the accuracy by the given test set.if the model predicts the correct output, which means our model is accurate.
```

```python
Types of Supervised Learning
1) Classification
2) Regression
```

```python
Classification

Classification algorithms are used when the output variable is categorical, which means there are two classes such as
Yes-No, True-False, Male-Female etc and spam Filtering

Classification algorithms are:
1) Random Forests
2) Decision Trees
3) Support Vector Machines
4) Logistic Regression
```

```python
Regression

Regression algorithms are used if the relation between input variable and output variable. it is used for the prediction of continuous variables
such as Weather Forecast, Stock trades etc.

Regression algorithms are:
1) Linear Regression
2) Non-Linear Regression
3) Bayesian Linear Regression
4) Polynomial Regression
5) Regression Trees
```

```python
Advantages of Supervised Learning

1) with the help of supervised Learning, the model can predict the output based on the prior experiences.
2) In Supervised Learning, we can have the exact idea about the classes of the objects.
3) It helps us to solve the real world problems like Fraud-Detection and spam filtering etc.

Disadvantages of Supervised Learning
1) Its not possible for handling complex tasks.
2) It cannot the predict the correct output if the test dataset from the training dataset.
3) Training Required lots of computation times.
4) In this model, we need enough knowledge about the classes of the objects.
```

```python
Unsupervised Learning
Unsupervised Learning is the learning method which the machine Learns without any Supervison.
that means machine learns without labels, categorical or classified and algorithms need to act on the data without any supervision.
Unsupervised Learning also known as unlabelled data.

Unsupervised Learning is also classified into two types. They are:
1) Clustering
2) Association
```

```python
The Goal of Unsupervised Learning is to find the underlying structures of the dataset, group that data according to the similarities
and represents the data in the compressed format.

why we use Unsupervised Learning ?

1) Unsupervised Learning is helpful for finding the insights from the data.
2) Unsupervised Learning is much similar as a human learns by their own experiences, which makes it closer to the real AI.
3) Unsupervised Learning is works on uncategorical and unlabelled data which makes this Unsupervised Learning is important.
4) In Real-world, we do not have input data with the corresponding output so to solve such cases, we need unsupervised Learning.

Working of Unsupervised Learning:

1) First we need to take unlabelled data which means it is not categorical and corresponding outputs also not given.
2) Now this unlabelled input data is fed to the machine learning model in order to train it.
3) Firstly, it will interept the raw data to find hidden patterns from the data and then will apply suitable algorithms such as
    k-Nearest Neighbors and decision trees etc.
```

```python
Types of Unsupervised Learning

1) Clustering
2) Association
```

```python
Clustering

Clustering is a method of grouping the objects into clusters such that objects with most similarities into a group and has less or no similarities
with the objects into a group. cluster analysis finds the commonalities between the data objects and categorized them as the presence and absence
of those commonalities.
```

```python
Association

Association Rule is a Unsupervised Learning method which is used for to find relationships between variables in a large dataset.
it determines the set of items that occur together in the dataset. it makes marketing strategy more effective. such as people who buy x item 
are also tend to purchase Y item. A typical example for association is Market Basket Analysis.

Example : For Suppose we ordered Biriyani in swiggy and also for chilling we ordered some cool Drink.
            x=Biriyani
            y=Limca Drink Bottle
```

```python
Unsupervised Learning Algorithms:

1) k-means clustering
2) KNN(k-nearest neighbors)
3) Principle Component Analysis
4) Independent Component Analysis
5) Apriori Algorithm
6) Hierarchical Clustering
7) Neural Networks
8) Singular value Decomposition
9) Anomaly Detection
```

```python
Advantages of Unsupervised Learning

1) Unsupervised Learning is used for more complex problems as compared to supervised Learning because in unsupervised learning, we don't use 
    labelled data
2) it is preferable as it is easy to get unlabelled data in comparision to labelled data.

Disadvantages of Unsupervised Learning

1) Unsupervised Learning is intrinsically more difficult than supervised learning as it does not have corresponding output.
2) The Result of unsupervised learning algorithm might be less accurate as input data is not labelled and algorithms do not know the exact output
    in advance.
```
