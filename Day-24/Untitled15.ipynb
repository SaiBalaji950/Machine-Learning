```python
ECLAT

ECLAT stands for Equivalence Class Transformations

Eclat is a mining frequent itemsets in transactional databases.
Eclat is different from both Apriori and FP Growth Algorithms. Because we use in both
Apriori and FP Growth algorithms we use horizontal format. In Eclat Algorithms we use
Vertical format. Each item is associated with a list of TID's.

TIDs (Transaction ID's.') 
```

```python
How Eclat Works

Vertical Data Representation:

The dataset is transformed into a vertical format, where each item points to a list of transaction IDs (TIDs).

TID Set Intersection:

For each item, the algorithm calculates the intersection of its TID set with those of other items to determine frequent itemsets.
If the size of the intersection exceeds a specified minimum support threshold, the itemset is considered frequent.

Recursive Depth-First Search:

Eclat employs a depth-first search approach, recursively combining items and 
calculating TID set intersections until all possible itemsets are explored.
```

```python
import pandas as pd
def eclat(transactions, min_support):
    item_tid_map = {}
    for tid, transaction in enumerate(transactions):
        for item in transaction:
            if item not in item_tid_map:
                item_tid_map[item] = set()
            item_tid_map[item].add(tid)
    frequent_itemsets = []
    def eclat_recursive(prefix, items):
        if not items:
            return
        for i in range(len(items)):
            item = items[i]
            tidset = item_tid_map[item]
            if len(tidset) >= min_support:
                frequent_itemsets.append((prefix + [item], len(tidset)))
                eclat_recursive(prefix + [item], items[i+1:])
    eclat_recursive([], list(item_tid_map.keys()))
    return frequent_itemsets
data = pd.read_csv('diabetes.csv')
transactions = data.values.tolist()
min_support = 0.003 * len(transactions)
frequent_itemsets = eclat(transactions, min_support)
print("Frequent Itemsets:")
for itemset in frequent_itemsets:
    print(f"Itemset: {itemset[0]}, Support Count: {itemset[1]}")
```
