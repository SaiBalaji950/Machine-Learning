```python
Affinity Propagation

Affinity Propogation is a clustering algorithm which identifies the exemplars (representative data points) in the dataset.
It is a type of clustering algorithm that does not require a pre-specified number of clusters, 
making it a useful tool for exploratory data analysis.
```

```python
import pandas as pd
a=pd.read_csv('diabetes.csv')
display(a)
```

```python
x=a.iloc[:,[3,6]].values
display(x)
```

```python
from sklearn.cluster import AffinityPropagation
b=AffinityPropagation(preference=50)
y_pred=b.fit_predict(x)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=100,c='red',label='Cluster1')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=100,c='blue',label='Cluster2')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=100,c='darkblue',label='Cluster3')
plt.scatter(x[y_pred==3,0],x[y_pred==3,1],s=100,c='orange',label='Cluster4')
plt.scatter(x[y_pred==4,0],x[y_pred==4,1],s=100,c='violet',label='Cluster5')
plt.title('Affinity Propagation')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```
