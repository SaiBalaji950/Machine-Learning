```python
Expectation Maximization Clustering using GMM (Gaussian Mixture Model)

Expectation Maximization Clustering Using Gaussian Mixture Model(GMM) is a Probabilistic model that 
assumes the data is generated from the mixture of several Gaussian Distributions.

It aims to find the parameters of these distributions that best fit the data.
```

```python
import pandas as pd
a=pd.read_csv('diabetes.csv')
display(a)
```

```python
x=a.iloc[:,[3,6]].values
display(x)
```

```python
from sklearn.mixture import GaussianMixture
b=GaussianMixture(n_components=3,random_state=42)
y_pred=b.fit_predict(x)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=100,c='red',label='Cluster1')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=100,c='blue',label='Cluster2')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=100,c='green',label='Cluster3')
plt.scatter(x[y_pred==3,0],x[y_pred==3,1],s=100,c='violet',label='Cluster4')
plt.scatter(x[y_pred==4,0],x[y_pred==4,1],s=100,c='red',label='Cluster5')
plt.title('Gaussian Mixture Models')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
data = pd.read_csv('diabetes.csv')
x = data[['Glucose', 'Insulin']].values 
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(x)
y_pred = gmm.predict(x)
plt.scatter(x[:, 0], x[:, 1], c=y_pred, s=40, cmap='viridis')
plt.title('GMM Clustering')
plt.xlabel('G')
plt.ylabel('Feature 2')
plt.show()
```
