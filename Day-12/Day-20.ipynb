```python
K-NN algorithm

1) K-NN stands for K-Nearest Neighbour algorithm.
2) K-Neaerest Neighbour is one of the simple algorithm based on supervised learning technique.
3) It assumes the similarities between new case/data and available cases and put the new data into category that is most similar to the available
    category.
4) It stores all the available data and classifies a new data based on similarity. this means when new data appears then it easily classified 
    into well-suitable category by using K-NN Algorithm.
5) It is used for classification problems as well as regression problems but mainly used on classification problems.
6) K-NN algorithm is a non-parametric algorithm, that means which don't make any assumptions and underlying data.
7) It is also a lazy-learner algorithm because it doesnot learn from the dataset immediately instead it stores the dataset and at the time of 
    classification and it performs action on the dataset.
8) KNN algorithm in the training phase, just it stores the dataset and when it gets new data and then classifies the data into category which is similar
    to the new data.
```

```python
Why do we need a K-NN algorithm ?

Suppose we have an image of creature that looks similar to cat and dog. We want to know the image whether it is cat or dog. so for this identification
purpose we need K-NN algorithm. Based on the identification, K-NN algorithm classifies the image it is dog or cat. Hence we need K-NN algorithm.
```

```python
How does K-NN work ?

1) Select the K of the neighbours. Here K refers to the number of the nearest neighbours.
2) Calculate the Euclidean Distance of K of the neighbors.
3) Consider the K-Nearest Neighbors as per calculated Euclidean Distance.
4) Among these k neighbors, count the number of data points in the each category.
5) Assign the new data points to the category for which the number of the neighbors is maximum.

Formula for Euclidean Distance:

Euclidean Distance of A and B is square root of ((x2-x1)^2+(y2-y1)^2)
```

```python
How to select the value of k in K-NN algorithm ?

1) There is no paricular way to determine the best value for k, so we need to try some values to find the best of the them.
    The Most prefered value for k is 5.
2) in Low values for k is such as k=0 or k=1 can be noisy and affect the outliers in the model.
3) Larger values are fine for k value, but it may find some difficulties.
```

```python
Advantages of K-NN algorithm

1) It is very easy to implement.
2) It is robust to the noisy training data.
3) It can be more affective if the training dataset is too large.
```

```python
Disadvantages of K-NN algorithm

1) Always need to determine the k value which may be complex some time.
2) The computation cost is too high because of calculating the distance for all the training samples.
```

```python
Implementation of K-NN algorithm by using python
step-1: Data PreProcessing
step-2: Fitting the K-NN algorithm to the Training set
step-3: Predicting the test result
step-4: Test accuracy of the result(Creation of Confusion matrix)
step-5: Visualizing the test set result.
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
a=pd.read_csv('data.csv')
display(a)
x=a[['id']]
y=a['perimeter_mean']
y_binned = pd.cut(y, bins=3, labels=[0, 1, 2])
x_train,x_test,y_train,y_test=train_test_split(x,y_binned,test_size=0.6,random_state=10)
b=StandardScaler()
x_train=b.fit_transform(x_train)
x_test=b.fit_transform(x_test)
print("x_train:\n",x_train)
print("x_test:\n",x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
c=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
c.fit(x_train,y_train)
y_pred=c.predict(x_test)
d=accuracy_score(y_test,y_pred)
e=confusion_matrix(y_test,y_pred)
f=classification_report(y_test,y_pred,zero_division=0)
print(d)
print(e)
print(f)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
a=pd.read_csv('data.csv')
x=a[['id']]
y=a['perimeter_mean']
y_binned=pd.cut(y,bins=3,labels=[0,1,2])
x_train,x_test,y_train,y_test=train_test_split(x,y_binned,test_size=0.2,random_state=42)
pipeline=Pipeline([('Scaler',StandardScaler()),('KNN',KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=3))])
pipeline.fit(x_train,y_train)
b=np.linspace(x.iloc[:,0].min() -1,x.iloc[:,0].max() +1,500)
b_grid=pd.DataFrame(b,columns=['id'])
z=pipeline.predict_proba(b_grid)[:,1]
plt.scatter(b, z, color='blue', label='Probability')
plt.scatter(x.iloc[:, 0], y_binned, c='red', marker='o', edgecolor='black', label='Data points')
plt.colorbar(label='probability')
plt.title("K-NN Algorithm")
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()
```

```python
Naive-Bayes

1) Naive Bayes is one of the supervised learning technique, which is based on bayes theorem and used for solving classification problem.
2) it is used for text classification that includes high-dimensional training dataset.
3) It is one of simple and most effective classification algorithms which helps in building the fast machine learning models that can make 
    quick predictions.
4) It is a Probabilistic classifier, which means it predicts on the basis of the probability of an object.
5) some popular examples of naive bayes algorithm are spam filtration, sentimental analysis and classifying articles.
```

```python
why it is called Naive-Bayes ?

Naive Bayes is derived into two words naive and bayes.

Naive: It assumes the occurence of certain features is independent of the occurence of other features.
Bayes: It is depend on the principle of Bayes theorem.
```

```python
Bayes Theorem:

Bayes Theorem is also known as Bayes rule or Bayes law which is used to determine the probability of the hypothesis with prior knowledge.
It depends on the conditional probability and the formula for Bayes theorem is

p(A|B)=((p(B|A).p(A))/p(B))

p(A|B)=Posterior Probability, Probability of hypothesis A on the observed event B.
p(B|A)=Likelihood Probability, Probability of the evidence given that the probability of a hypothesis is True.
p(A)=Prior Probability,Probability of hypothesis before observing the evidence.
P(B)=Marginal Probability, Probability of evidence.

where hypothesis means is a proposed explanation that can be tested and evidence is the gathering information whether hypothesis is correct or wrong.
```
