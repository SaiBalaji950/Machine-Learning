```python
Random Forest

Random Forest is a most popular Machine Learning Algorithm belongs to Supervised Learning Technique and used for both classification and 
Regression Problems. It is based on the concept of the Ensemble Learning, which means it combines the multiple classifiers to solve 
complex problem and it improves the performance of the model.

As a name suggests, Random Forest is a classifier that contains number of decision trees on a various subsets of the given dataset and 
takes the average to improve the predictive accuracy of the dataset. Instead of relying on one decision tree, the random forest takes the
prediction from each tree and based on the majority votes of predictions, and it predicts the final output.

The Greatest Number of Trees leads to the higher accuracy and prevents the problem of overfitting.
```

```python
Assumptions of the Random Forest

Since the random forest combines multiple trees to predict the class of the dataset, it is possible that some decision trees 
may predict correct output and some decision decision trees may predict wrong output. but together it predicts correct output.
there are two assumptions of Random Forest. they are

1) There should be some actual values in the feature variable of the dataset so that the classifier can predict accurate results 
    rather than guessed result.
2) The Predictions from much tree must have very low correlations.
```

```python
Why we use Random Forest ?

It takes less training as compared to other algorithms.
It predicts output with Higher Accuracy, even for the large dataset it runs efficiently.
It can also maintain when a large proportion of data was missing.
```

```python
Working of Random Forest

1) Select Random k data points from the training set.
2) Build the decision trees associated with the selected data points(subsets).
3) Choose the number N for decision trees that you want to build.
4) Repeat Steps 1 and 2.
5) For new data points, find the predictions of each decision tree, and assign the new data points to the category that wins the
    majority votes.
```

```python
Applications of Random Forest

1) Banking
2) Marketing
3) Land Use
4) Medicine
```

```python
Advantages of Random Forest

1) Random Forest have a capable to perform both Classification and Regression Tasks.
2) It is a capable of handling the large datasets with high dimensionality.
3) It enhances the accuracy of the model and prevents from the overfitting issue.
```

```python
Disadvantages of Random Forest

Although Random Foresgt have capable to perform both classification and Regression Tasks but it majorly it is used for Regression tasks.
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=pd.read_csv('diabetes.csv')
x=a.iloc[:,[2,3]]
y=a.iloc[:,4]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.6,random_state=38)
b=StandardScaler()
x_train=b.fit_transform(x_train)
x_test=b.fit_transform(x_test)
c=RandomForestClassifier(n_estimators=10,criterion='entropy')
c.fit(x_train,y_train)
y_pred=c.predict(x_test)
x_pred=c.predict(x_train)
d=accuracy_score(y_test,y_pred)
e=confusion_matrix(y_test,y_pred)
f=classification_report(y_test,y_pred,zero_division=0)
print("Accuracy:\n",d)
print("confusion_matrix:\n",e)
print("Classification_Report:\n",f)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=pd.read_csv('diabetes.csv')
x=a[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
y=a['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
b=StandardScaler()
x_train=b.fit_transform(x_train)
x_test=b.fit_transform(x_test)
c=RandomForestClassifier(n_estimators=10,criterion='entropy')
c.fit(x_train,y_train)
y_pred=c.predict(x_test)
x_pred=c.predict(x_train)
d=accuracy_score(y_test,y_pred)
e=confusion_matrix(y_test,y_pred)
f=classification_report(y_test,y_pred,zero_division=0)
print("Accuracy:\n",d)
print("Confusion_Matrix:\n",e)
print("Classification_Report:\n",f)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
a=pd.read_csv('diabetes.csv')
x=a[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
y=a['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
pipeline=Pipeline([('scaler',StandardScaler()),('RFC',RandomForestClassifier(n_estimators=10,criterion='entropy'))])
pipeline=pipeline.fit(x_train,y_train)
b=np.linspace(x.iloc[:,0].min() -1,x.iloc[:,1].max() +1,500)
b_grid=pd.DataFrame(np.repeat(b[:,np.newaxis],x.shape[1],axis=1),columns=x.columns)
c=pipeline.predict_proba(b_grid)[:,1]
plt.scatter(b,c,color='g',label='predicted probability')
plt.scatter(x.iloc[:,0],y,c=y,cmap='Reds',marker='o',edgecolor='b',label='data points')
plt.colorbar(label='Probability')
plt.title("Random Forest")
plt.xlabel('All Data')
plt.ylabel("Outcome")
plt.show()
```
