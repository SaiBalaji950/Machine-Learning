```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=pd.read_csv('diabetes.csv')
x=a[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
y=a['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
b=StandardScaler()
x_train=b.fit_transform(x_train)
x_test=b.fit_transform(x_test)
c=RandomForestClassifier(n_estimators=10,criterion='entropy')
c.fit(x_train,y_train)
y_pred=c.predict(x_test)
x_pred=c.predict(x_train)
d=accuracy_score(y_test,y_pred)
e=confusion_matrix(y_test,y_pred)
f=classification_report(y_test,y_pred)
print("Accuracy:\n",d)
print("confusion matrix:\n",e)
print("classification report:\n",f)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=pd.read_csv('diabetes.csv')
x=a[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
y=a['Outcome']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
b=StandardScaler()
x_train=b.fit_transform(x_train)
x_test=b.fit_transform(x_test)
c=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
c.fit(x_train,y_train)
y_pred=c.predict(x_test)
x_pred=c.predict(x_train)
d=accuracy_score(y_test,y_pred)
e=confusion_matrix(y_test,y_pred)
f=classification_report(y_test,y_pred)
print("Accuracy:\n",d)
print("Confusion Matrix:\n",e)
print("Classification Report:\n",f)
```
