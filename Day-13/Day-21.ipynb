```python
Naive Bayes 

1) Naive Bayes is one of the Supervised Learning Algorithm and based on Bayes theorem and used for to solve classification problems.
2) It is used for text classification which includes high dimensionality training datasets.
3) Naive Bayes are one of tthe most and effective classification algorithms which helps in building the fast machine learning models which can make 
    fast predictions. 
4) It is also a Probability classifier, which means it predicts basis of the probability of an object.
5) some popular exampes of the naive bayes algorithms are spam filtration,sentimental analysis and classifying articles.
```

```python
why it is called naive bayes ?

Naive bayes can be derived into two words naive and bayes.

Naive: it assumes the occurence of certain features is independent of the occurence of another features.
Bayes: it is based on principle of bayes theorem.
```

```python
Bayes Theorem

Bayes theorem is also known as Bayes rule or Bayes law which is used to determine the probability of hypothesis with prior knowledge.
it depends on the conditional probability and the formula of bayes theorem is:

P(A|B)=((P(B|A).P(A))/P(B))

P(A|B)=Posterior Probability, Probability on hypothesis A on the observed event on B.
P(B|A)=Likelihood Probability, Probability of the evidence that gives the probability of hypothesis is True.
P(A)=Prior Probability, Probability of hypothesis before observing the evidence.
p(B)=Marginal Probability, Probability of evidence

where hypothesis is a proposed explanation that can be explained and evidence means it checks the hypothesis is correct or wrong.
```

```python
Advatages of Naive Bayes

1) Naive Bayes is fast and simplest machine learning algorithms to predict a class of datasets.
2) It is used for binary classification as well as multi-class classifications.
3) It performs well in multi-class classification while compared to other machine learning algorithms.
4) It is the most popular choice for Text classification.

Disadvantages of Naive Bayes

It assumes that all features are independent or unrelated, so it cannot learn the relationship between features.
```

```python
Applications of Naive Bayes

1) It is used for credit scoring.
2) It is used in medical data classification.
3) It can be used in real world predictions because naive bayes is an eager learner.
4) It is used in text classification such as spam filtration and sentimental analysis.
```

```python
Implementation of Naive Bayes using Python

Steps to implement:
step-1: Data Pre-processing step
step-2: Fitting Naive Bayes to the Training set
step-3: Predicting the test result
step-4: Test accuracy of the result(Creation of Confusion matrix)
step-5: Visualizing the test set result.
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
a=pd.read_csv('data.csv')
display(a)
x=a[['id']]
y=a['perimeter_mean']
y_binned=pd.cut(y,bins=3,labels=[0,1,2])
x_train,x_test,y_train,y_test=train_test_split(x,y_binned,test_size=0.3,random_state=42)
b=StandardScaler()
x_train=b.fit_transform(x_train)
x_test=b.fit_transform(x_test)
print("x_train:\n",x_train)
print("x_test:\n",x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
c=GaussianNB()
c.fit(x_train,y_train)
y_pred=c.predict(x_test)
x_pred=c.predict(x_train)
d=accuracy_score(y_test,y_pred)
e=confusion_matrix(y_test,y_pred)
f=classification_report(y_test,y_pred,zero_division=0)
print("Accuracy:",d)
print("Confusion_matrix:\n",e)
print("classification_report:\n",f)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
a=pd.read_csv('data.csv')
x=a[['id']]
y=a['perimeter_mean']
y_binned=pd.cut(y,bins=3,labels=[0,1,2])
x_train,x_test,y_train,y_test=train_test_split(x,y_binned,test_size=0.3,random_state=42)
pipeline=Pipeline([('scaler',StandardScaler()),('Naive_Bayes',GaussianNB())])
#pipeline = Pipeline([('scaler', StandardScaler()),('Naive_Bayes', GaussianNB())])
pipeline=pipeline.fit(x_train,y_train)
b=np.linspace(x.iloc[:,0].min() -1,x.iloc[:,0].max() +1,500)
b_grid=pd.DataFrame(b,columns=['id'])
z=pipeline.predict_proba(b_grid)[:,1]
plt.scatter(b,z,color='r',label='predicted probability')
plt.scatter(x.iloc[:,0],y_binned,c=y_binned,cmap='winter',marker='o',edgecolor='k',label='data points')
plt.colorbar(label='probability')
plt.title('Naive Bayes')
plt.xlabel('id')
plt.ylabel('perimeter_names')
plt.legend()
plt.show()
```
